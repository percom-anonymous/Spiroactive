```python
import matplotlib.pyplot as plt
from tueplots import bundles
import numpy as np
import pickle
```

```python
from tueplots import figsizes
figsizes.tmlr2023(nrows=2)["figure.figsize"]
plt.rcParams.update(bundles.tmlr2023(rel_width=0.4,nrows=1.5))
```

```python
## Load data
with open("../results/figure7/pid_removed_fev1.pkl","rb") as f:
    pid_removed_fev1 = pickle.load(f)

with open("../results/figure7/pid_removed_fvc.pkl","rb") as f:
    pid_removed_fvc = pickle.load(f)
```

```python
#count common number of points removed in every iteration
countt=[]
for i in range(len(pid_removed_fvc)):
    countt.append(len(set(pid_removed_fvc[:i+1]).intersection(set(pid_removed_fev1[:i+1]))))

print(countt)

x = [i for i in range(0, 33)]
plt.rcParams['axes.labelsize']=11
plt.rcParams['legend.fontsize']=10
plt.rcParams['axes.titlesize']=12
plt.rcParams['xtick.labelsize']=10
plt.rcParams['ytick.labelsize']=10
x= [i for i in range(1,34)]

plt.plot(x,countt,marker="")
plt.plot(x,x,color='red',linestyle='--',marker="")
plt.xlabel("Number of pool points")
plt.ylabel("Number of common points")

plt.savefig("../figures/figure7.pdf",bbox_inches='tight')
```
